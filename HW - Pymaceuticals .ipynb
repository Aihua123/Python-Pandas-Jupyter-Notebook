```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

trial_file = 'data/clinicaltrial_data.csv'
drug_file = 'data/mouse_drug_data.csv'
```

```python
df_trial = pd.read_csv(trial_file)
df_drug = pd.read_csv(drug_file)
merged_df = df_trial.merge(df_drug, how='left', on='Mouse ID')
merged_df.head()
```

# Tumor Response to Treatment

```python
df_volume_mean = merged_df.groupby(['Drug','Timepoint']).agg({'Tumor Volume (mm3)':'mean'}).reset_index()
df_volume_mean.head(10)
```

```python
df_volume_sem = merged_df.groupby(['Drug','Timepoint']).agg({'Tumor Volume (mm3)':'sem'}).reset_index()
df_volume_sem.head(10)
```

```python
df_mean = df_volume_mean.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
df_mean
```

```python
df_sem = df_volume_sem.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
df_sem
```

```python
x = df_sem.index

plt.errorbar(x, df_mean['Capomulin'],yerr = df_sem['Capomulin'],marker='o',linewidth=1,markersize=5)
plt.errorbar(x, df_mean['Infubinol'],yerr = df_sem['Infubinol'],marker='s',linewidth=1,markersize=5)
plt.errorbar(x, df_mean['Ketapril'],yerr = df_sem['Ketapril'],marker='*',linewidth=1,markersize=5)
plt.errorbar(x, df_mean['Placebo'],yerr = df_sem['Placebo'],marker='d',linewidth=1,markersize=5)

plt.title("Tumor Response to Treatment")
plt.xlabel("Time(Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc="upper left")
plt.grid()
```

# Metastatic Response to Treatment

```python
df_met_mean = merged_df.groupby(['Drug','Timepoint']).agg({'Metastatic Sites':'mean'}).reset_index()
df_met_mean.head()
```

```python
df_met_sem = merged_df.groupby(['Drug','Timepoint']).agg({'Metastatic Sites':'sem'}).reset_index()
df_met_sem.head()
```

```python
df_mean_1 = df_met_mean.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
df_mean_1
```

```python
df_sem_1 = df_met_sem.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
df_sem_1
```

```python
plt.errorbar(x, df_mean_1['Capomulin'],yerr = df_sem_1['Capomulin'],marker='o',linestyle=':',linewidth=1,markersize=5)
plt.errorbar(x, df_mean_1['Infubinol'],yerr = df_sem_1['Infubinol'],marker='s',linestyle=':',linewidth=1,markersize=5)
plt.errorbar(x, df_mean_1['Ketapril'],yerr = df_sem_1['Ketapril'],marker='*',linestyle=':',linewidth=1,markersize=5)
plt.errorbar(x, df_mean_1['Placebo'],yerr = df_sem_1['Placebo'],marker='d',linestyle=':',linewidth=1,markersize=5)

plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration(Days)")
plt.ylabel("Met.Sites")
plt.legend(loc="upper left")
plt.grid()
```

# Survival Rates

```python
df_mouse_count = merged_df.groupby(['Drug','Timepoint']).agg({'Mouse ID':'count'}).reset_index()
df_mouse_count = df_mouse_count.rename(columns={"Mouse ID": "Number of Mouse"})
df_mouse_count.head()
```

```python
df_count = df_mouse_count.pivot(index='Timepoint', columns='Drug', values='Number of Mouse')
df_count
```

```python
c_pct = df_count['Capomulin']/df_count.ix[0,'Capomulin']*100
i_pct = df_count['Infubinol']/df_count.ix[0,'Infubinol']*100
k_pct = df_count['Ketapril']/df_count.ix[0,'Ketapril']*100
p_pct = df_count['Placebo']/df_count.ix[0,'Placebo']*100

plt.plot(x, c_pct, marker='o', label = 'Capomulin')
plt.plot(x, i_pct, marker='s', label = 'Infubinol')
plt.plot(x, k_pct, marker='v', label = 'Ketapril')
plt.plot(x, p_pct, marker='d', label = 'Placebo')

plt.legend(loc='lower left')
plt.title("Survival During Treatment")
plt.xlabel("Time(Days)")
plt.ylabel("Survival Rate(%)")
plt.grid()
plt.savefig("Survival During Treatment.png")
```

# Summary Bar Graph

```python
total_pct_changes = (df_mean.loc[45]/df_mean.loc[0]-1)*100
total_pct_changes
```

```python
plt.figure(figsize=(8.4, 5.6))
g = total_pct_changes[['Capomulin','Infubinol','Ketapril','Placebo']].plot(kind='bar',color=['green', 'red', 'red', 'red'],
                                                                       width=1, grid=True, align='edge')

g.set_yticks([-20, 0, 20, 40, 60])
g.set_title('Tumor Change Over 45 Day Treatment')
g.set_ylabel('% Tumor Volume Change')
g.set_xticklabels(['Capomulin','Infubinol','Ketapril','Placebo'], rotation=0, fontsize=12,ha='left')


rects = g.patches
labels = ["{:.0f}%".format(i)for i in total_pct_changes[['Capomulin','Infubinol','Ketapril','Placebo']]]

for rect, label in zip(rects, labels):
    if label == "{:.0f}%".format(total_pct_changes['Capomulin']):
        g.text(rect.get_x() + rect.get_width() / 2, rect.get_height()+12, label,
               fontsize=14, color='white', ha='center')
    else:
        g.text(rect.get_x() + rect.get_width() / 2, rect.get_height()-8, label,
               fontsize=14, color='white', ha='center')

plt.savefig("Tumor Change Over 45 Day Treatment.png")
```

# Three observable trends based on the data

1.	Among the four treatments (Capomulin, Infubinol, Ketapril, and a placebo), the tumor volume declined over the course of 45 days under the Capomulin treatment while the tumor volume had a similarly significant increase under the other three treatments.


2.	The number of metastatic (cancer spreading) sites under each of the four treatments increased over the course of 45 days. However, the Capomulin treatment had the fewest metastatic sites at the end of the course.


3.	The Capomulin treatment has the highest survival rate, over 80% at the end of the course. The survival rates under the other three treatments had dropped below 50% after 45 days.
